```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("tested.csv")  

# Basic view
df.head()
```

```python
# Summary statistics
print(df.describe())

# Info about columns & missing values
print(df.info())

# Value counts for a categorical column
print(df['Sex'].value_counts())
```

```python
# Pairplot (for numerical relationships)
sns.pairplot(df, hue="Survived")
plt.show()

# Heatmap of correlations
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

Observations from Pairplot Visualization

Passenger Class (Pclass):
 Passengers in 1st class had the highest survival rate.
 3rd class passengers had the lowest chance of survival.

Age:
Children under 10 had a significantly higher survival rate compared to adults.
Elderly passengers (above ~60) had a much lower survival probability.

Fare:
Higher ticket fares are associated with a greater chance of survival.
 Many low-fare passengers did not survive, especially in 3rd class.

SibSp (Siblings/Spouses Aboard):
Survival rates were better for passengers traveling with 12 family members.
Very high SibSp values (large families) had lower survival chances.

Parch (Parents/Children Aboard):
Moderate family presence (13 Parch) increased chances of survival.
Passengers traveling alone (Parch = 0) had lower odds.

Combined Factors:
Survival likelihood increases when Pclass is high, Fare is high, and Age is low.
Clusters of survivors are clearly visible in the 1st class, high-fare, younger age group.

Observations from Correlation Heatmap

Pclass & Fare:
There is a strong negative correlation between Pclass and Fare  higher-class passengers paid more for tickets.

Pclass & Survival:
A negative correlation exists between Pclass and Survived, meaning passengers in higher classes had better survival rates.

Fare & Survival:
A positive correlation shows that higher fares are linked to higher survival chances, reflecting socio-economic advantages.

Parch & SibSp:
Moderate positive correlation indicates that passengers with more parents/children often also traveled with siblings/spouses.

Age:
Weak correlation with survival  age alone is not a strong predictor, but combined with class and fare, it influences survival.

Key Insight:
The most important predictors of survival appear to be Fare and Pclass, supported by a smaller influence from family size (Parch + SibSp).

```python
# Survival rate by gender
sns.barplot(x="Sex", y="Survived", data=df)
plt.show()

# Survival rate by passenger class
sns.barplot(x="Pclass", y="Survived", data=df)
plt.show()
```

Observation: Survival Rate by Gender

Females had a significantly higher survival rate compared to males.
This trend supports the "women and children first" evacuation policy followed during the Titanic disaster.
Gender appears to be a strong predictor of survival probability.

Observation: Survival Rate by Passenger Class

Passengers in 1st class had the highest survival rate, indicating better access to lifeboats and rescue efforts.
Survival rates were significantly lower for 2nd and 3rd class passengers, possibly due to physical barriers, location of cabins, or socio-economic prioritization.
Passenger class is a strong indicator of survival chances, showing a clear socio-economic impact on safety during the disaster.

```python
# Histogram of Age
sns.histplot(df['Age'], kde=True)
plt.show()

# Boxplot of Fare by Class
sns.boxplot(x="Pclass", y="Fare", data=df)
plt.show()

# Scatterplot of Age vs Fare
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=df)
plt.show()
```

1. Age Distribution (Histogram with KDE)

The age distribution is right-skewed, with most passengers in the 2030 years age range.
There are fewer passengers in the elderly group (60+), and some very young passengers (under 5 years) are also present.
This suggests the Titanic had a relatively young adult demographic onboard.

2. Fare Distribution by Passenger Class (Boxplot)

First-class passengers paid the highest fares, with a wide range and several high-value outliers (above 500).
Second-class fares are much lower and more consistent, while third-class fares are the lowest and most concentrated.
This highlights the economic disparity among classes, which may correlate with survival chances.

3. Age vs Fare by Survival (Scatterplot)

Higher fares are generally associated with better survival rates, as seen in the upper price range where more survivors are present.
Passengers of all ages are represented, but younger passengers with higher fares have a greater survival proportion.
Many passengers who paid lower fares did not survive, particularly in the third-class group

Summary of Findings

Survival and Gender
Females had a much higher survival rate than males, indicating a likely application of the "women and children first" evacuation policy.

Survival and Passenger Class
First-class passengers had the highest survival rates, while second and third classes had lower rates, showing a strong link between socio-economic status and survival.

Age Distribution
The passenger age distribution is right-skewed, with most passengers between 2030 years old.
Few elderly passengers were onboard, and there were a small number of children.

Fare and Class
First-class fares were significantly higher and varied widely, with notable outliers.
Third-class fares were the lowest and most concentrated, reinforcing economic disparities on the ship.

Age vs Fare vs Survival
Passengers paying higher fares generally had better survival odds, especially among younger and middle-aged groups.
Lower fare-paying passengers had more fatalities, especially in third class.

Correlation Insights
Survival is positively correlated with fare and passenger class (lower Pclass number = higher survival).
Age had a weaker correlation, suggesting survival was influenced more by economic status and gender than by age alone.
